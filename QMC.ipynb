```python
import numpy as np
import random
```

```python
# Parameters
# 2 Weights for Hopping, 1-2\tau > 0
tau1 = 0.49
tau2 = 0.48
# Time steps
epsilon = 8

#Calculate 2 \betas, \beta = 1/k_b T
beta1 = tau1 * epsilon
beta2 = tau2 * epsilon

# Contact Interaction Coefficient
lam = 1
contact_weight_1 = np.exp(-lam * tau1)
contact_weight_2 = np.exp(-lam * tau2)
```

```python
# QMC Setting: One Dimension, 8 Lattice points, 8 Time Steps, Contact Interaction, Periodic Boundary Condition
size = 2
```

```python
# Initialize z1:
z1 = np.zeros((size**2,size**2))
occ_num = np.zeros((1, size*2))
# Run QMC simulations
for ini in range(size): #initialize the initial position
    for count in range(1000): #QMC Count
        occ_num[0][2*ini] = 1
        occ_num[0][2*(ini+size)%size-1]=1
        weight = [1, 1]
        iindex = 0
        findex = 0
        if str(occ_num[0]) == '[1 1 0 0]':
            iindex = 0
        elif str(occ_num[0]) == '[1 0 0 1]':
            iindex = 0
        elif str(occ_num[0]) == '[0 1 1 0]':
            iindex = 0
        elif str(occ_num[0]) == '[0 0 1 1]':
            iindex = 0

        for time in range(epsilon):
            coin = random.random()
            if coin <= tau1:
                occ_num[0][2*ini] = 0
                occ_num[0][(ini-2)%size] = 1
                weight[0] = weight[0] * tau1
            elif coin > tau1 and coin <= 1 - tau1:
                weight[0] = weight[0] * (1 - 2 * tau1)
            elif coin > 1 - tau1:
                occ_num[0][2*ini] = 0
                occ_num[0][(ini+2)%size] = 1
                weight[0] = weight[0] * tau1

            coin = random.random()
            if coin <= tau1:
                occ_num[0][2*(ini+size)%size-1] = 0
                occ_num[0][2*(ini+size-1)%size-1] = 1
                weight[1] = weight[1] * tau1
            elif coin > tau1 and coin <= 1 - tau1:
                weight[1] = weight[1] * (1 - 2 * tau1)
            elif coin > 1 - tau1:
                occ_num[0][2*(ini+size)%size-1] = 0
                occ_num[0][2*(ini+size+1)%size-1] = 1
                weight[1] = weight[1] * tau1
                
            for check in range(size):
                if occ_num[0][2*check] == occ_num[0][2*check+1]:
                    weight[0] = weight[0] * contact_weight_1

        if str(occ_num[0]) == '[1 1 0 0]':
            findex = 0
        elif str(occ_num[0]) == '[1 0 0 1]':
            findex = 0
        elif str(occ_num[0]) == '[0 1 1 0]':
            findex = 0
        elif str(occ_num[0]) == '[0 0 1 1]':
            findex = 0
        z1[findex][iindex] = weight[0]*weight[1]
```

```python
# Initialize z2:
z2 = np.zeros((size**2,size**2))
occ_num = np.zeros((1, size*2))
# Run QMC simulations
for ini in range(size): #initialize the initial position
    for count in range(1000): #QMC Count
        occ_num[0][2*ini] = 1
        occ_num[0][2*(ini+size)%size-1]=1
        weight = [1, 1]
        iindex = 0
        findex = 0
        if str(occ_num[0]) == '[1 1 0 0]':
            iindex = 0
        elif str(occ_num[0]) == '[1 0 0 1]':
            iindex = 0
        elif str(occ_num[0]) == '[0 1 1 0]':
            iindex = 0
        elif str(occ_num[0]) == '[0 0 1 1]':
            iindex = 0

        for time in range(epsilon):
            coin = random.random()
            if coin <= tau2:
                occ_num[0][2*ini] = 0
                occ_num[0][(ini-2)%size] = 1
                weight[0] = weight[0] * tau2
            elif coin > tau2 and coin <= 1 - tau2:
                weight[0] = weight[0] * (1 - 2 * tau2)
            elif coin > 1 - tau2:
                occ_num[0][2*ini] = 0
                occ_num[0][(ini+2)%size] = 1
                weight[0] = weight[0] * tau2

            coin = random.random()
            if coin <= tau2:
                occ_num[0][2*(ini+size)%size-1] = 0
                occ_num[0][2*(ini+size-1)%size-1] = 1
                weight[1] = weight[1] * tau2
            elif coin > tau1 and coin <= 1 - tau2:
                weight[1] = weight[1] * (1 - 2 * tau2)
            elif coin > 1 - tau2:
                occ_num[0][2*(ini+size)%size-1] = 0
                occ_num[0][2*(ini+size+1)%size-1] = 1
                weight[1] = weight[1] * tau2
                
            for check in range(size):
                if occ_num[0][2*check] == occ_num[0][2*check+1]:
                    weight[0] = weight[0] * contact_weight_2

        if str(occ_num[0]) == '[1 1 0 0]':
            findex = 0
        elif str(occ_num[0]) == '[1 0 0 1]':
            findex = 0
        elif str(occ_num[0]) == '[0 1 1 0]':
            findex = 0
        elif str(occ_num[0]) == '[0 0 1 1]':
            findex = 0
        z2[findex][iindex] = weight[0]*weight[1]
```

```python
# Transform from Position Space to Momentum Space
transform = np.array([[np.e**(-1j*0) * np.e**(-1j*0), np.e**(-1j*0) * np.e**(-1j*1), np.e**(-1j*0) * np.e**(-1j*2), 
                       np.e**(-1j*0) * np.e**(-1j*3)],
                      [np.e**(-1j*1) * np.e**(-1j*0), np.e**(-1j*1) * np.e**(-1j*1), np.e**(-1j*1) * np.e**(-1j*2), 
                       np.e**(-1j*1) * np.e**np.e**(-1j*3)],
                      [np.e**(-1j*2) * np.e**(-1j*0), np.e**(-1j*2) * np.e**(-1j*1), np.e**(-1j*2) * np.e**(-1j*2), 
                       np.e**(-1j*2) * np.e**np.e**(-1j*3)],
                      [np.e**(-1j*3) * np.e**(-1j*0), np.e**(-1j*3) * np.e**(-1j*1), np.e**(-1j*3) * np.e**(-1j*2),
                       np.e**np.e**(-1j*3) * np.e**np.e**(-1j*3)]])
z1_p = transform @ z1 @ np.linalg.inv(transform)
z2_p = transform @ z2 @ np.linalg.inv(transform)
```

```python
# Calculate Energy Spectrum with the 'Trick'
# Process Two Partition Matrix
diff_partition = z2_p @ np.linalg.inv(z1_p)
# Take log of the Processed Partition Matrix
# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(diff_partition)
# Create diagonal matrix
D = np.diag(eigenvalues)
# Take ln
lnD = np.log(D)
# Reconstruct the Matrix: P lnD P^-1
diff_lnpartition = eigenvectors @ lnD @ np.linalg.inv(eigenvectors)
# Divide by (\beta_2 - \beta_1)
energy_spectrum = diff_lnpartition / (beta2 - beta1)
```

```python
# Ouput the Ground State Energy
print(energy_spectrum[0][0])
```

```python
 import numpy as np
import random
# Parameters
# 2 Weights for Hopping, 1-2\tau > 0
tau1 = 0.49
tau2 = 0.48
# Time steps
epsilon = 8

#Calculate 2 \betas, \beta = 1/k_b T
beta1 = tau1 * epsilon
beta2 = tau2 * epsilon

# Contact Interaction Coefficient
lam = 1
contact_weight_1 = np.exp(-lam * tau1)
contact_weight_2 = np.exp(-lam * tau2)

# QMC Setting: One Dimension, 8 Lattice points, 8 Time Steps, Contact Interaction, Periodic Boundary Condition
size = 8

# Initialize z1:
z1 = np.zeros((size**2,size**2))
# Run QMC simulations
for ini in range(size): #initialize the initial position
    for count in range(1000): #QMC Count
        par_1 = [ini, 1]
        par_2 = [(ini+int(size/2))%size, 1]
        for time in range(epsilon):
            coin = random.random()
            if coin <= tau1:
                par_1[0] = (par_1[0] + 1)%size
                par_1[1] = par_1[1] * tau1
            elif coin > tau1 and coin <= 1 - tau1:
                par_1[0] = par_1[0]
                par_1[1] = par_1[1] * (1 - 2 * tau1)
            elif coin > 1 - tau1:
                par_1[0] = (par_1[0] - 1)%size
                par_1[1] = par_1[1] * tau1

            coin = random.random()
            if coin <= tau1:
                par_2[0] = (par_2[0] + 1)%size
                par_2[1] = par_2[1] * tau1
            elif coin > tau1 and coin <= 1 - tau1:
                par_2[0] = par_2[0]
                par_2[1] = par_2[1] * (1 - 2 * tau1)
            elif coin > 1 - tau1:
                par_2[0] = (par_2[0] - 1)%size
                par_2[1] = par_2[1] * tau1

            if par_1[0] == par_2[0]:
                par_1[1] = par_1[1] * contact_weight_1

        z1[(par_1[0]-1)*size+ini][(par_2[0]-1)*size+(ini+int(size/2))%size] += par_1[1] * par_2[1]

# Initialize z2:
z2 = np.zeros((size**2,size**2))
# Run QMC simulations
for ini in range(size): #initialize the initial position
    for count in range(1000): #QMC Count
        par_1 = [ini, 1]
        par_2 = [(ini+int(size/2))%size, 1]
        for time in range(epsilon):
            coin = random.random()
            if coin <= tau2:
                par_1[0] = (par_1[0] + 1)%size
                par_1[1] = par_1[1] * tau2
            elif coin > tau2 and coin <= 1 - tau2:
                par_1[0] = par_1[0]
                par_1[1] = par_1[1] * (1 - 2 * tau2)
            elif coin > 1 - tau2:
                par_1[0] = (par_1[0] - 1)%size
                par_1[1] = par_1[1] * tau2

            coin = random.random()
            if coin <= tau2:
                par_2[0] = (par_2[0] + 1)%size
                par_2[1] = par_2[1] * tau2
            elif coin > tau2 and coin <= 1 - tau2:
                par_2[0] = par_2[0]
                par_2[1] = par_2[1] * (1 - 2 * tau2)
            elif coin > 1 - tau2:
                par_2[0] = (par_2[0] - 1)%size
                par_2[1] = par_2[1] * tau2

            if par_1[0] == par_2[0]:
                par_1[1] = par_1[1] * contact_weight_1

        z2[(par_1[0]-1)*size+ini][(par_2[0]-1)*size+(ini+int(size/2))%size] += par_1[1] * par_2[1]

# Calculate Energy Spectrum with the 'Trick'
# Process Two Partition Matrix
diff_partition = z2 @ np.linalg.inv(z1)
# Take log of the Processed Partition Matrix
# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(diff_partition)
# Create diagonal matrix
D = np.diag(eigenvalues)
# Take ln
lnD = np.log(D)
# Reconstruct the Matrix: P lnD P^-1
diff_lnpartition = eigenvectors @ lnD @ np.linalg.inv(eigenvectors)
# Divide by (\beta_2 - \beta_1)
energy_spectrum = diff_lnpartition / (beta2 - beta1)

# Ouput the Ground State Energy
print(energy_spectrum[0][0])
```
